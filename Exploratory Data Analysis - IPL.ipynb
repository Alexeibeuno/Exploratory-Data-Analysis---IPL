# Batsman Statistics

```python
deliveries['Cumulative_batsman_runs'] = deliveries.groupby(['match_id','batsman'])['batsman_runs'].cumsum()
deliveries['ballcount'] = 1
deliveries['cummulative_ball_count'] = deliveries.groupby(['match_id','batsman'])['ballcount'].cumsum()
deliveries['Fours'] = deliveries.batsman_runs.apply(lambda x:1 if x==4 else 0)
deliveries['Sixes'] = deliveries.batsman_runs.apply(lambda x:1 if x==6 else 0)
deliveries['No Ball'] = deliveries.noball_runs.apply(lambda x:1 if x>0 else  0)
deliveries['Wide Ball'] = deliveries.wide_runs.apply(lambda x:1 if x>0 else 0)
deliveries['Maiden Overs'] = deliveries.total_runs.apply(lambda x:1 if x==0 else 0)
deliveries['Bounds'] = deliveries.total_runs.apply(lambda x: 1 if x==4 or x==6 else 0)
deliveries['Matches_played'] = deliveries.groupby(['batsman'])['match_id'].cumsum()
```

```python
batsman_stats = (pd.concat([
    deliveries.batsman.value_counts().rename("Total Balls Faced"),
    deliveries.non_striker.value_counts().rename("Balls at Non-Stricker's End"),
    deliveries.groupby(["batsman"])["batsman_runs"].sum().rename("Total Runs Scored"),
    deliveries.groupby(["batsman"])["Cumulative_batsman_runs"].max().rename("Maximum Runs in a Match"),
    deliveries.groupby(["batsman"])["Fours"].sum().rename("Fours"),
    deliveries.groupby(["batsman"])["Sixes"].sum().rename("Sixes"),
    (deliveries.groupby(["inning","batsman"])["batsman_runs"].sum().rename("Highest Score").reset_index()
     .groupby(["batsman"])["Highest Score"].max().rename("Highest Score per Inning")),
     (deliveries[deliveries['Cumulative_batsman_runs']>=100].groupby(['batsman']).size().rename('Centuries')),
    (deliveries[deliveries['Cumulative_batsman_runs']>=50].groupby(['batsman']).size().rename('Half Centuries')),
    (deliveries[deliveries['cummulative_ball_count']==6].groupby(['batsman']).size().rename('Run Rate')),
    (deliveries[deliveries['Matches_played']>30].groupby(["match_id","batsman"])["Bounds"].sum().rename("Boundaries per match")
     .reset_index().groupby(["batsman"])["Boundaries per match"].mean().rename("Average Boundaries per match"))
     ],
       axis=1,
    )
    .fillna(0)
    .reset_index()
    .rename(columns={"index":"Batsman"})
                ) 
batsman_stats.sort_values(by='Total Runs Scored',ascending=False)[:5]
```

Batsman Strick Rate

```python
batsman_stats['Strike Rate'] = (batsman_stats['Total Runs Scored']/batsman_stats['Total Balls Faced'])*100
batsman_stats
```

# Bowler Statistics

Top 5 "Tidy Bowlers"

```python
bowler_stats = (pd.concat([
   deliveries.groupby(["bowler"])["over"].sum().rename("Overs"),
   deliveries.bowler.value_counts().rename("Total Balls Bowled"),
   deliveries.groupby(["bowler"])['Maiden Overs'].sum().rename("Maiden overs"),
   (deliveries[deliveries["dismissal_kind"].isin(["caught","bowled","lbw","stumped","caught and bowled","hit wicket"])]
   .groupby(["bowler"]).size().rename("Total Wickets")),
    deliveries.groupby(["bowler"])['No Ball'].sum().rename("No Balls"),
    deliveries.groupby(["bowler"])['Wide Ball'].sum().rename("Wide Balls"),
    (deliveries.groupby(["match_id","bowler"])["total_runs"].sum().rename("Total runs conceded").reset_index()
     .groupby(["bowler"])['Total runs conceded'].sum().rename("Total runs conceded per match")),
    (deliveries[deliveries["dismissal_kind"].isin(["caught","bowled","lbw","stumped","caught and bowled","hit wicket"])]
    .groupby(['match_id','bowler']).size().rename("Total Wickets").reset_index()
    .groupby(['bowler'])['Total Wickets'].mean().rename("Average Wickets taken per match"))
],
    axis=1,
)
.fillna(0)
.reset_index()
.rename(columns={"index":"Bowler"}))
bowler_stats.sort_values(by='Total Wickets',ascending=False)[:5]
```
